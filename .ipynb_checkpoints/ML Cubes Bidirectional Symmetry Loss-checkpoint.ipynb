```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class QuaternionLinear(nn.Module):
    def __init__(self, in_features, out_features):
        super(QuaternionLinear, self).__init__()
        # Each weight has 4 components: r, i, j, k
        self.r = nn.Linear(in_features, out_features)
        self.i = nn.Linear(in_features, out_features)
        self.j = nn.Linear(in_features, out_features)
        self.k = nn.Linear(in_features, out_features)

    def forward(self, x):
        # x shape: (batch, in_features, 4)
        xr, xi, xj, xk = x[..., 0], x[..., 1], x[..., 2], x[..., 3]

        # Quaternion multiplication rules
        yr = self.r(xr) - self.i(xi) - self.j(xj) - self.k(xk)
        yi = self.r(xi) + self.i(xr) + self.j(xk) - self.k(xj)
        yj = self.r(xj) - self.i(xk) + self.j(xr) + self.k(xi)
        yk = self.r(xk) + self.i(xj) - self.j(xi) + self.k(xr)

        y = torch.stack([yr, yi, yj, yk], dim=-1)
        return y


class QuaternionMLP(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(QuaternionMLP, self).__init__()
        self.fc1 = QuaternionLinear(input_dim, hidden_dim)
        self.fc2 = QuaternionLinear(hidden_dim, hidden_dim)
        self.fc3 = QuaternionLinear(hidden_dim, hidden_dim)
        self.fc4 = QuaternionLinear(hidden_dim, output_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        return x
```
