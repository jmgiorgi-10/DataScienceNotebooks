```python
import pandas as pd
import numpy as np

LAG = 5

df = pd.read_csv("ABB.csv")
print(df.columns)

features = ["Open", "High", "Low", "Volume", "Close"]
df = df[features]
# limit number of datapoints.
df = df[0:1000]

print(df.columns)

y = df["Close"]
print(y)
```

```python
# Start with linear regression --> need to introduce lag to adapt regression to time-series data.
def lag_features(data, lag=10):
    
    df = data.copy()
    for feature in features:
        for i in range(1, lag + 1):
            df[f'{feature}_lag{i}'] = df[feature].shift(i)

    for feature in features:
        df.pop(feature) # now remove non-lagged features, to avoid data leakage.

    # remove nans due to lagged features
    df = df.dropna()
        
    return df
    # print("Lagged dataframe: ", df)
```

```python
# # Confirm time-series lagging worked.
# import matplotlib.pyplot as plt
# plt.plot(df['Open'][0:10])
# plt.show()
# plt.plot(X['Open_lag1'][0:10])
# plt.show()
```

```python
# Define train/test split ratio
train_size = int(len(df) * 0.3)

# Split the data into training and test sets based on time
train, test = df[:train_size], df[train_size:]
test = df.reset_index(drop='True')

y_train = train['Close'][LAG:]
X_train = lag_features(train, LAG)

y_test = test['Close'][LAG:]
X_test = lag_features(test, LAG)

print(y_test)
# print(X_train[0:10])
# print(y_train[0:10])
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
# print(y_pred['time'])
plt.plot(y_pred, label = 'predicted')
plt.plot(y_test, label = 'actual')
plt.title("Predicted vs Actual ABB stock price over time")
plt.xlabel("Time step")
plt.ylabel("Closing stock price")
plt.legend()
plt.show()

# plt.plot(y_test)
# plt.title("Actual ABB stock price over time")
# plt.xlabel("Time step")
# plt.ylabel("Closing stock price")
# plt.show()
```
